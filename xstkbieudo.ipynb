```python
import matplotlib.pyplot as plt
import numpy as np
data = [162,157,174,158,170,160,169,166,153,172,171,164,165,156]
mean, sigma = np.mean(data), np.std(data)
index = []
y_mean = [mean]*14
for i in range(len(data)):
    index.append(i+1)
ax = plt.subplot()
mean_line = ax.plot(index,y_mean, label="Mean", linestyle = "--",color ="black")
plt.plot(index,data, 'go-',color="red")
plt.title("Biểu đồ chiều cao")
plt.xlabel("Số lần đo")
plt.ylabel("Chiều cao")
legend = ax.legend(loc='upper right')
plt.show()
```
